## Multivariate Linear Regression

```python
class Linear:
    
    #l.predict
    def predict (self,X, theta):
        """
        X - np array (m, n+1)
        theta - np array (n+1, 1)
        """
    
        return np.dot(X, theta)


    #------------------------------->
    def error(self, X, y, theta):
        """
        params:
            X - np array (m, n+1)
            y - np array(m,1)
            theta - np array(n+1, 1)
        return :
            scalar value = loss value
        """
    
        y_hat = self.predict(X, theta)
        err = np.mean((y_hat - y) **2)
        #can also do sum

        return err
    
    def gradient(self, X, y, theta):
        """
            X - np array (m, n+1)
            y - np array(m,1)
            theta - np array(n+1, 1)
        """
    
        y_hat = self.predict(X, theta)
        grad = np.dot(X.T, (y_hat-y))


        return grad/X.shape[0]
        #grad/m
    
    def fit(self, X, y, lr=0.1, max_itr =100):
    
        theta = np.zeros((X.shape[1],1))
    #     theta (n+1,1)
        error_list = []

        for i in range(max_itr):
            err = self.error(X, y, theta)
            error_list.append(err)

            grad = self.gradient(X, y, theta)

            #updation Rule
            theta = theta -lr*(grad)
        return (theta, error_list)
```

```python
l = Linear()
```

```python
opt_theta, error_list = l.fit(X, y)
```
